# 1. Introduction to Pandas

Pandas is a powerful open-source library that's essential for working with data in Python. It provides data structures and functions that make handling structured data efficient and straightforward.

While you might be familiar with Python's built-in data structures like lists, Pandas introduces its own structures, primarily `Series` and `DataFrames`, which are specifically designed for working with structured data and offer significant advantages for data analysis. Unlike basic lists, Pandas data structures have features like labeled indexes, making data alignment and manipulation much easier. They also come with a wide array of built-in functions optimized for data operations.

### 1.1 Series

A Pandas `Series` is like a single column of data. It can hold different types of information, like numbers or text, and each item has a label, called its index. Think of it as a list with an index that doesn't have to be just numbers starting from zero.

Let's look at a simple example of creating a Series from a Python list.

```python
import pandas as pd

# Creating a Series from a list
data = [10, 20, 30, 40, 50]
s = pd.Series(data)
print("Series from a list:")
print(s)
```

You can also create a Series and specify your own labels for the index instead of using the default numbers.

```python
# Creating a Series with a custom index
data_with_index = [100, 200, 300]
index_labels = ['a', 'b', 'c']
s_indexed = pd.Series(data_with_index, index=index_labels)
print("\nSeries with a custom index:")
print(s_indexed)
```

### 1.2 DataFrame

A Pandas `DataFrame` is a two-dimensional table with rows and columns. It's great for organizing data where each column can hold a different type of data (like numbers, text, etc.). Unlike a simple list of lists, a DataFrame has clear column names and row indexes, which makes working with your data much more intuitive.

Let's see how to create a simple DataFrame, often from a Python dictionary where keys become column names.

```python
# Creating a DataFrame from a dictionary
data = {'col1': [1, 2, 3, 4],
        'col2': ['A', 'B', 'C', 'D']}
df = pd.DataFrame(data)
print("DataFrame from a dictionary:")
print(df)
```

You can also specify both the row index and column names when creating a DataFrame.

```python
# Creating a DataFrame with a custom index and columns
data_complex = {'col_a': [5, 6],
                'col_b': [7, 8]}
index_labels = ['row1', 'row2']
column_labels = ['col_a', 'col_b']
df_complex = pd.DataFrame(data_complex, index=index_labels, columns=column_labels)
print("\nDataFrame with custom index and columns:")
print(df_complex)
```

### 1.4 Accessing and Selecting Data

Once you have a DataFrame, you'll often need to grab specific pieces of data from it. Pandas gives you flexible ways to do this, primarily using `.loc[]` and `.iloc[]`. Understanding the difference between them is key!

*   `.loc[]` is used for **label-based** indexing. This means you use the *names* or *labels* of your rows and columns to select data. If your DataFrame has a custom index (like names or dates instead of numbers), `.loc[]` is what you'll use to refer to those specific labels.
*   `.iloc[]` is used for **integer-based** indexing. This means you use the *position* of the rows and columns to select data, starting from 0, just like with Python lists. It doesn't matter what the labels are; `.iloc[]` only cares about the numerical position.

```python
# Let's create a DataFrame with a custom index and more data to better see the difference
data_custom_index = {'col_A': [10, 20, 30, 40, 50],
                     'col_B': [60, 70, 80, 90, 100],
                     'col_C': [110, 120, 130, 140, 150]}
custom_index = ['row_alpha', 'row_beta', 'row_gamma', 'row_delta', 'row_epsilon']
df_custom = pd.DataFrame(data_custom_index, index=custom_index)

print("DataFrame with custom index:")
print(df_custom)
```

You can select a single column by putting its name in square brackets. This works the same whether you have a default or custom index.

```python
# Selecting a single column using column name
print("\nSelecting 'col_A':")
print(df_custom['col_A'])
```

To select multiple columns, put a list of column names in the square brackets.

```python
# Selecting multiple columns using a list of column names
print("\nSelecting 'col_A' and 'col_B':")
print(df_custom[['col_A', 'col_B']])
```

Using `.loc[]`, you select a row by its *label*. See how we use `'row_beta'` which is the actual label in our custom index?

```python
# Selecting a row using .loc[] (by index label)
print("\nSelecting row with label 'row_beta' using .loc[]:")
print(df_custom.loc['row_beta'])
```

Using `.iloc[]`, you select a row by its *integer position*. Even though the label is `'row_beta'`, its position is 1 (since we start counting from 0).

```python
# Selecting a row using .iloc[] (by integer position)
print("\nSelecting row at integer position 1 using .iloc[]:")
print(df_custom.iloc[1])
```

You can select a specific value by providing both the row *label* and column *label* to `.loc[]`.

```python
# Selecting a specific value using .loc[]
print("\nSelecting value at row label 'row_alpha' and column label 'col_B' using .loc[]:")
print(df_custom.loc['row_alpha', 'col_B'])
```

Similarly, use `.iloc[]` with integer *positions* for both row and column.

```python
# Selecting a specific value using .iloc[]
print("\nSelecting value at row integer position 0, column integer position 1 using .iloc[]:")
print(df_custom.iloc[0, 1])
```

You can also slice rows using *labels* with `.loc[]`. Remember that with `.loc[]`, the end label is *included* in the slice. So `'row_alpha':'row_gamma'` includes rows with those labels and everything in between.

```python
# Slicing rows using .loc[]
print("\nSlicing rows from label 'row_alpha' to 'row_gamma' (inclusive) using .loc[]:")
print(df_custom.loc['row_alpha':'row_gamma'])
```

Slicing with `.iloc[]` uses integer *positions*, and the end position is *excluded*, just like standard Python slicing. So `0:3` includes rows at positions 0, 1, and 2, but not 3.

```python
# Slicing rows using .iloc[]
print("\nSlicing rows from integer position 0 to 3 (exclusive) using .iloc[]:")
print(df_custom.iloc[0:3])
```

### 1.5 Filtering Data

Filtering data means picking out only the rows that meet certain conditions. This is super useful for finding specific information in your dataset.

```python
# Filtering rows where the value in 'col1' is greater than 2
print("Original DataFrame:")
print(df)
print("\nFiltering rows where 'col1' > 2:")
print(df[df['col1'] > 2])
```

You can also combine multiple conditions using logical operators like `&` (and) and `|` (or). Make sure to put each condition in parentheses!

```python
# Filtering with multiple conditions
print("\nFiltering rows where 'col1' > 2 AND 'col2' is 'C':")
print(df[(df['col1'] > 2) & (df['col2'] == 'C')])
```

### 1.6 Adding and Dropping Columns

You'll often need to add new information as a column or remove columns you don't need.

```python
# Adding a new column named 'col3'
df['col3'] = [True, False, True, False]
print("DataFrame after adding 'col3':")
print(df)
```

To remove a column, you can use the `.drop()` method. You need to specify the column name and `axis=1` to tell Pandas you're dropping a column ( `axis=0` is for dropping rows).

```python
# Dropping a column
df_dropped = df.drop('col3', axis=1) # axis=1 specifies dropping a column
print("\nDataFrame after dropping 'col3':")
print(df_dropped)

# Keep in mind that .drop() usually gives you a *new* DataFrame with the column removed.
# If you want to change the original DataFrame directly, you can add `inplace=True`
# like this: df.drop('col3', axis=1, inplace=True)
```

### 4.1 Basic Plotting

Let's start with a simple line plot using some sample data to understand the basic structure of creating a plot.

Before we plot, we often create a figure and axes to control the appearance of our plot.

*   `plt.figure()`: Creates a new figure (an empty canvas for your plot). You can specify the size using the `figsize` parameter (width, height in inches).
    *   [Matplotlib `figure` Documentation](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.figure.html)
*   `plt.plot()`: Draws points with lines connecting them. This is the standard function for creating line plots.
    *   [Matplotlib `plot` Documentation](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.plot.html)
*   `plt.title()`: Adds a title to your plot.
    *   [Matplotlib `title` Documentation](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.title.html)
*   `plt.xlabel()`: Adds a label to the x-axis.
    *   [Matplotlib `xlabel` Documentation](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.xlabel.html)
*   `plt.ylabel()`: Adds a label to the y-axis.
    *   [Matplotlib `ylabel` Documentation](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.ylabel.html)
*   `plt.grid()`: Adds a grid to the plot, which can help with reading values.
    *   [Matplotlib `grid` Documentation](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.grid.html)
*   `plt.show()`: Displays the plot. Without this, the plot might not appear depending on the environment.
    *   [Matplotlib `show` Documentation](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.show.html)

```python
# Sample data
x = np.linspace(0, 10, 100) # 100 points between 0 and 10
y = np.sin(x)

# Create a simple line plot
plt.figure(figsize=(8, 4))
plt.plot(x, y)
plt.title("Simple Sine Wave")
plt.xlabel("X-axis")
plt.ylabel("Y-axis (sin(x))")
plt.grid(True)
plt.show()
```

You can easily plot data directly from a Pandas DataFrame. Let's use our `df_data` DataFrame. We'll look at how the `temp` column (temperature) changes over time, using the `datetime` column. We need to make sure the `datetime` column is actually treated as dates and times by Pandas, so we'll convert it using `pd.to_datetime()`.

*   `pd.to_datetime()`: Converts a column to datetime objects, which is essential for time series plotting.
    *   [Pandas `to_datetime` Documentation](https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.to_datetime.html)
*   `plt.xticks(rotation=...)`: Rotates the labels on the x-axis, useful for long labels like dates.
    *   [Matplotlib `xticks` Documentation](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.xticks.html)
*   `plt.tight_layout()`: Automatically adjusts plot parameters for a tight layout, preventing labels from overlapping.
    *   [Matplotlib `tight_layout` Documentation](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.tight_layout.html)

```python
# Plotting 'temp' over 'datetime' from df_data
# Convert 'datetime' to a proper datetime format first for better plotting
df_data['datetime'] = pd.to_datetime(df_data['datetime'])

plt.figure(figsize=(12, 6))
plt.plot(df_data['datetime'], df_data['temp'])
plt.title("Temperature Over Time")
plt.xlabel("Date and Time")
plt.ylabel("Temperature (°C)")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

### 4.2 Different Plot Types

Matplotlib supports many different types of plots. Let's look at a few common ones and how to create them with our bike-sharing data. We can treat columns like 'season', 'holiday', 'workingday', and 'weather' as categorical variables for plotting purposes, even though they contain numbers.

**Scatter Plot:** Useful for showing the relationship between two numerical variables. Each point on the plot represents an observation. Let's see if there's a relationship between the actual temperature (`temp`) and the 'feeling' temperature (`atemp`).

*   `plt.scatter()`: Creates a scatter plot. You provide the data for the x and y axes. The `alpha` parameter controls the transparency of the points (0 is fully transparent, 1 is fully opaque), which is useful when you have many overlapping points.
    *   [Matplotlib `scatter` Documentation](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.scatter.html)

```python
# Scatter plot of 'temp' vs 'atemp'
plt.figure(figsize=(8, 6))
plt.scatter(df_data['temp'], df_data['atemp'], alpha=0.5)
plt.title("Feeling Temperature vs. Actual Temperature")
plt.xlabel("Actual Temperature (°C)")
plt.ylabel("Feeling Temperature (°C)")
plt.grid(True)
plt.show()
```

**Histogram:** Shows the distribution of a single numerical variable. It divides the data into bins and shows how many observations fall into each bin. Let's look at the distribution of `humidity`.

*   `plt.hist()`: Creates a histogram. The `bins` parameter determines the number of bins (bars) in the histogram. `edgecolor` sets the color of the edges of the bars.
    *   [Matplotlib `hist` Documentation](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.hist.html)
*   `plt.grid(axis='y')`: Adds a grid only along the y-axis. You can use `axis='x'` for the x-axis grid.
    *   [Matplotlib `grid` Documentation](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.grid.html)

```python
# Histogram of 'humidity'
plt.figure(figsize=(8, 6))
plt.hist(df_data['humidity'], bins=30, edgecolor='black')
plt.title("Distribution of Humidity")
plt.xlabel("Humidity (%)")
plt.ylabel("Frequency")
plt.grid(axis='y')
plt.show()
```

**Bar Plot:** Good for comparing a numerical value across different categories. It shows bars whose lengths are proportional to the values they represent. Let's see the average `temp` for each `season`. We'll need to calculate the average temperature per season first using Pandas' `.groupby()` and `.mean()`.

*   `.groupby()`: A Pandas method used to group rows of a DataFrame based on the values in a specified column.
    *   [Pandas `groupby` Documentation](https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.groupby.html)
*   `.mean()`: A Pandas method often used after `groupby()` to calculate the average of a numerical column within each group.
    *   [Pandas `mean` Documentation](https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.core.groupby.GroupBy.mean.html)
*   `series.plot(kind='bar')`: Pandas Series and DataFrames have a `.plot()` method built on top of Matplotlib, making it easy to create plots directly from your data structures. `kind='bar'` specifies that you want a bar plot.
    *   [Pandas `.plot()` Documentation](https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.plot.html)

```python
# Calculate average temp per season
average_temp_per_season = df_data.groupby('season')['temp'].mean()

# Bar plot of average temp per season
plt.figure(figsize=(8, 6))
average_temp_per_season.plot(kind='bar', edgecolor='black')
plt.title("Average Temperature per Season")
plt.xlabel("Season (1: Winter, 2: Spring, 3: Summer, 4: Fall)")
plt.ylabel("Average Temperature (°C)")
plt.xticks(rotation=0)
plt.grid(axis='y')
plt.show()
```

**Box Plot:** Useful for visualizing the distribution of a numerical variable across different categories. A box plot shows the median, quartiles, and potential outliers for each category. Let's look at the distribution of `windspeed` for each `weather` type.

*   `df.boxplot()`: Pandas DataFrames have a built-in `.boxplot()` method that uses Matplotlib to create box plots. You specify the column for the numerical data (`column`) and the column for the categories (`by`).
    *   [Pandas `.boxplot()` Documentation](https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.boxplot.html)

```python
# Box plot of windspeed by weather type
plt.figure(figsize=(10, 6))
df_data.boxplot(column='windspeed', by='weather')
plt.title("Windspeed Distribution by Weather Type")
plt.xlabel("Weather Type") # Add a note explaining weather codes if needed
plt.ylabel("Windspeed")
plt.suptitle('') # Suppress the default title added by pandas boxplot
plt.show()
```

```python
# ### ADD YOUR CODE HERE###
# 1. Scatter Plot: 'temp' vs 'registered'
plt.figure(figsize=(8, 6))
plt.scatter(df_data['temp'], df_data['registered'], alpha=0.5)
plt.title("Temperature vs. Registered Users")
plt.xlabel("Temperature (°C)")
plt.ylabel("Registered Users")
plt.grid(True)
plt.show()
# ### END OF YOUR CODE###
```

```python
# ### ADD YOUR CODE HERE###
# 2. Histogram: 'atemp' distribution
plt.figure(figsize=(8, 6))
plt.hist(df_data['atemp'], bins=25, edgecolor='black')
plt.title("Distribution of Feeling Temperature")
plt.xlabel("Feeling Temperature (°C)")
plt.ylabel("Frequency")
plt.grid(axis='y')
plt.show()
# ### END OF YOUR CODE###
```

```python
# ### ADD YOUR CODE HERE###
# 3. Bar Plot: Average windspeed per holiday type
average_windspeed_per_holiday = df_data.groupby('holiday')['windspeed'].mean()

plt.figure(figsize=(8, 6))
average_windspeed_per_holiday.plot(kind='bar', edgecolor='black')
plt.title("Average Windspeed per Holiday Type")
plt.xlabel("Holiday Type (0: Not, 1: Yes)")
plt.ylabel("Average Windspeed")
plt.xticks(rotation=0)
plt.grid(axis='y')
plt.show()
# ### END OF YOUR CODE###
```

```python
# ### ADD YOUR CODE HERE###
# 4. Box Plot: 'humidity' distribution by 'season'
plt.figure(figsize=(10, 6))
df_data.boxplot(column='humidity', by='season')
plt.title("Humidity Distribution by Season")
plt.xlabel("Season")
plt.ylabel("Humidity (%)")
plt.suptitle('')
plt.show()
# ### END OF YOUR CODE###
```

**Think about it:** What does the box plot of humidity by season tell you about how humidity varies across the different seasons in this dataset? Are there any seasons that seem to have a wider range of humidity values than others?

```python
average_temp_per_season = df_data.groupby('season')['temp'].mean()

plt.figure(figsize=(8, 6))
average_temp_per_season.plot(kind='bar', edgecolor='black')
plt.title("Average Temperature per Season")
plt.xlabel("Season (1: Winter, 2: Spring, 3: Summer, 4: Fall)")
plt.ylabel("Average Temperature (°C)")
plt.xticks(rotation=0)
plt.grid(axis='y')
plt.show()
```
